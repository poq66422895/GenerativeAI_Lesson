```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

我第一次嘗試的圖形是心型圖形，但我發現心型好像是蠻常見的函數圖形，網上有很多的實例，因此我就繼續找有哪些特別的函數圖形，然後就發現螺旋圖形了。這個程式碼不單單可以展示螺旋圖形，還可以調整其中的一些參數，讓螺旋圖形產生變化!

```python
def spiral_curve(a, b, n):
  t = np.linspace(0, 10*np.pi, n)  # 調整螺旋的大小
  x = a * np.exp(b * t) * np.cos(t)
  y = a * np.exp(b * t) * np.sin(t)
  return x, y

a = 0.4  # 調整螺旋的寬窄
b = 0.1  # 調整螺旋的旋轉速度
n = 1000  # 點的數量

x, y = spiral_curve(a, b, n)

plt.figure(figsize=(8, 8))  # 調整圖形的大小
plt.plot(x, y)
plt.title("Spiral")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
